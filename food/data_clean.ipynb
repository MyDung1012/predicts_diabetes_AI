```python
import re, math, warnings
import numpy as np
import pandas as pd
```

IMPORT DATA


```python
df = pd.read_csv('Food_and_Nutrition.csv')
df.head()
```

tạo cột BMI dựa trên giới tính, tuổi, chiều cao, cân nặng

```python
df["Height_m"] = df["Height"] / 100  # đổi cm sang m
df["BMI"] = df["Weight"] / (df["Height_m"] ** 2)
df["BMI"] = df["BMI"].round(1)
print(df["BMI"].describe())
```

Ma hoa cot Gender


```python
df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0})
df.head()
```

Tao cot heart_disease va hypertension


```python
dcol = "Disease"  #
hyper_terms = ["hypertension"]
heart_terms = ["heart disease"]

df["hypertension"] = pd.Series([np.nan]*len(df), dtype="Int64")
df["heart_disease"] = pd.Series([np.nan]*len(df), dtype="Int64")
for i, row in df.iterrows():
    disease = str(row[dcol]).lower()
    if any(term in disease for term in hyper_terms):
        df.at[i, "hypertension"] = 1
    else:
        df.at[i, "hypertension"] = 0
    if any(term in disease for term in heart_terms):
        df.at[i, "heart_disease"] = 1
    else:
        df.at[i, "heart_disease"] = 0
df.drop(columns=[dcol], inplace=True)
df.head()
```

```python
diet_col = "Dietary Preference"

def diet_encode(x):
    if pd.isna(x):
        return np.nan
    s = str(x).lower()
    veg_terms = ["vegetarian", "vegan","pescatarian"]  
    normal_terms = ["omnivore"]          
    if any(t in s for t in veg_terms):
        return 1
    if any(t in s for t in normal_terms):
        return 0
    return np.nan

df["Dietary Preference"] = df[diet_col].apply(diet_encode).astype("Int64")
df.head()
```

```python
# Xuất ra file CSV
OUTPUT_PATH = "food_data.csv"

# index=False để không ghi chỉ số hàng vào file CSV
df.to_csv(OUTPUT_PATH, index=False, encoding="utf-8")

print(f"Đã lưu file CSV tại: {OUTPUT_PATH}")
```
