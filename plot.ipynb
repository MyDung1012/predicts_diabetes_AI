```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Thiết lập style toàn cục
plt.style.use('seaborn-v0_8')  # Sử dụng style seaborn-v0_8
sns.set_context("notebook", font_scale=1.2)  # Tăng kích thước font
plt.rcParams['font.family'] = 'Arial'  # Font chữ hiện đại

# Load dữ liệu từ 3 mô hình
gan_history = pd.read_excel("GAN_LSTM_history.xlsx")
smotenc_history = pd.read_excel("SMOTENC_LSTM_history.xlsx")
tsom_history = pd.read_excel("TSOM_LSTM_history.xlsx")

# ------------------ Biểu đồ 1: Accuracy ------------------ #
plt.figure(figsize=(12, 6), dpi=100)  # Tăng độ phân giải

# GAN - xanh dương
plt.plot(gan_history['accuracy'], label='GAN Train', color='#1f77b4', linewidth=2.5, linestyle='-')
plt.plot(gan_history['val_accuracy'], label='GAN Val', color='#1f77b4', linewidth=2.5, linestyle='--', alpha=0.7)

# SMOTENC - cam
plt.plot(smotenc_history['accuracy'], label='SMOTENC Train', color='#ff7f0e', linewidth=2.5, linestyle='-')
plt.plot(smotenc_history['val_accuracy'], label='SMOTENC Val', color='#ff7f0e', linewidth=2.5, linestyle='--', alpha=0.7)


# Tùy chỉnh tiêu đề và nhãn
plt.title('LSTM Model Accuracy Comparison', fontsize=16, weight='bold', pad=15)
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)

# Tùy chỉnh lưới
plt.grid(True, linestyle='--', alpha=0.7)

# Tùy chỉnh chú thích - đặt bên ngoài biểu đồ, bên phải
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=10, frameon=True, shadow=True, facecolor='white', edgecolor='black')

# Tối ưu bố cục
plt.tight_layout()
plt.show()

# ------------------ Biểu đồ 2: Loss ------------------ #
plt.figure(figsize=(12, 6), dpi=100)

# GAN - xanh dương
plt.plot(gan_history['loss'], label='GAN Train', color='#1f77b4', linewidth=2.5, linestyle='-')
plt.plot(gan_history['val_loss'], label='GAN Val', color='#1f77b4', linewidth=2.5, linestyle='--', alpha=0.7)

# SMOTENC - cam
plt.plot(smotenc_history['loss'], label='SMOTENC Train', color='#ff7f0e', linewidth=2.5, linestyle='-')
plt.plot(smotenc_history['val_loss'], label='SMOTENC Val', color='#ff7f0e', linewidth=2.5, linestyle='--', alpha=0.7)



# Tùy chỉnh tiêu đề và nhãn
plt.title('LSTM Model Loss Comparison', fontsize=16, weight='bold', pad=15)
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Loss', fontsize=12)

# Tùy chỉnh lưới
plt.grid(True, linestyle='--', alpha=0.7)

# Tùy chỉnh chú thích - đặt bên ngoài biểu đồ, bên phải
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=10, frameon=True, shadow=True, facecolor='white', edgecolor='black')

# Tối ưu bố cục
plt.tight_layout()
plt.show()
```
